```python
for i, row in dat.iterrows():
    a = row.blood1_biobank
    if a is np.nan:
        continue
    if ';' in a:
        a = a.split(';')[1]
    exp = r'[C,S,P][0-9][0-9]'
    match = re.findall(r'[C,S,P][0-9][0-9]',a)
    match = ''.join(match)
    
    #serum
    serum =  re.findall(r'S[0-9][0-9]',match)
    serum = ''.join(serum)
    number = re.findall(r'[0-9][0-9]',serum)
    number = list(map(int, number))
    if len(number) > 0:
        dat.at[i,'blood1_serum'] = max(number)

    #plasma = 
    plasma =  re.findall(r'P[0-9][0-9]',match)
    plasma = ''.join(plasma)
    numberp = re.findall(r'[0-9][0-9]',plasma)
    numberp = list(map(int, numberp))
    if len(numberp) > 0:
        dat.at[i,'blood1_plasma'] = max(numberp)

    #cells =
    cells =  re.findall(r'C[0-9][0-9]',match)
    cells = ''.join(cells)
    numberc = re.findall(r'[0-9][0-9]',cells)
    numberc = list(map(int, numberc))
    if len(numberc) > 0:
        dat.at[i,'blood1_cell'] = max(numberc)
```
